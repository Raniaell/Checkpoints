#### Linear Regression

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn import metrics
#extract x and y
x=data["carat"].values[:,np.newaxis]
y=data["price"].values

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=40) #splitting data with test size of 20%
```

```python
model=LinearRegression()   #build linear regression model
model.fit(x_train,y_train)  #fitting the training data
```

```python
#Plot linear regression
plt.scatter(x,y,color="r")
plt.title("Linear Regression")
plt.ylabel("price")
plt.xlabel("carat")
plt.plot(x,model.predict(x),color="k")
plt.show()
```

```python
predicted=model.predict(x_test) #testing our modelâ€™s performance
print("MSE", mean_squared_error(y_test,predicted))
print("R squared", metrics.r2_score(y_test,predicted))
```

R squared indicates that the model is performing well with a value close to 85%.

```python
features=["carat","'x'","'y'","'z'"]
for f in features:
    # extract x and y from our data
    y=data["price"].values
    x=data[f].values.reshape(-1,1)
    # splitting the data with test size = 20%
    x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.20, random_state=40)
    model=LinearRegression() # build linear regression
    model.fit(x_train, y_train) # fitting the training set
    predicted=model.predict(x_test) # test our model's performance
    print("Linear Regression with", f)
    print("MSE:", mean_squared_error(y_test, predicted))
    print("R squared:", metrics.r2_score(y_test, predicted))
    print("-----------------------------------------------------")
```

All the models are capable of well explaining the model's performance but the best feature is 'carat' with the best R squared and a lesser MSE value.

#### Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
features=["carat","'x'","'y'","'z'"]
for f in features:
    # extract x and y from our data
    y=data["price"].values
    x=data[f].values.reshape(-1,1)
    # splitting the data with test size = 20%
    x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.20, random_state=40)
    lg=LinearRegression()
    print("\n")
    print("Polynomial Regression with",f,"\n")
    for i in range (2,11):
        poly=PolynomialFeatures(degree=i)
        # transformation to polynmial
        x_train_fit=poly.fit_transform(x_train)
        # fitting the model
        lg.fit(x_train_fit, y_train)
        # testing the model
        x_test_=poly.fit_transform(x_test)
        predicted=lg.predict(x_test_)
        print("Polynomial Degree :", i)
        print("MSE:", mean_squared_error(y_test, predicted))
        print("R squared:", metrics.r2_score(y_test, predicted))
```

Using the polynomial regression, there's an improvement in the model's performance. Also with the variation with the polynomial degree there are better R squared results for example for the feature Carat the best values of R squared and MSE are obtained for degree 8.

```python
#Let's try to plot the polynomial regression of the feature 'carat'
from sklearn.preprocessing import PolynomialFeatures
x= data["carat"].values.reshape(-1,1)
y= data["price"].values
poly = PolynomialFeatures(degree = 8) 
x_poly = poly.fit_transform(x) 
poly.fit(x_poly, y) 
lg=LinearRegression()
lg.fit(x_poly, y) 

plt.scatter(x, y, color="r")
plt.title("Polynomial regression of price vs carat with polynomial degree 8 ")
plt.ylabel("price")
plt.xlabel("carat")
plt.plot(x, lg.predict(poly.fit_transform(x)), color="k")
```

#### K nearest neighbor

```python
data1=data 
data1=data1.drop('depth',axis=1)
data1=data1.drop('table',axis=1)

from sklearn.preprocessing import LabelEncoder
Cut= LabelEncoder()
Color= LabelEncoder()
Clarity= LabelEncoder()

data1['cut_n']= Cut.fit_transform(data1['cut'])
data1['color_n']= Color.fit_transform(data1['color'])
data1['clarity_n']= Clarity.fit_transform(data1['clarity'])
data1
```

Label encodings for: cut column is 1: b'good, 2: b'ideal, 3: b'premium, 4: b'very good, 5: b'fair;

for color column is 0: b'D, 1: b'E, 2: b'F, 3: b'G 4: b'H, 5: b'I, 6: b'J;

for clarity column is 0: b'I1, 1: b'IF, 2: b'SI1, 3: b'SI2, 4: b'VS1, 5: b'VS2, 6: b'VVS1, 7: b'VVS2.

```python
data1=data1.drop('cut',axis=1)
data1=data1.drop('color',axis=1)
data1=data1.drop('clarity',axis=1)
data1
```

```python
#Choosing the features as x and the target as y
x=data1[["carat","'x'","'y'","'z'","color_n","cut_n","clarity_n"]]
y=data1['price']
data1
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.2, random_state=40)

from sklearn.neighbors import KNeighborsRegressor
knn_model = KNeighborsRegressor(n_neighbors=3)

knn_model.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
train_preds = knn_model.predict(x_train) #evaluate the prediction error on the training data (the result should be relatively good)
mse = mean_squared_error(y_train, train_preds)
rmse = sqrt(mse)
print("the rmse result for the training data is :", rmse)
#RMSD is always non-negative, and a value of 0 (almost never achieved in practice) would indicate a perfect fit to the data. 
#In general, a lower RMSD is better than a higher one
```

```python
test_preds = knn_model.predict(x_test) #evaluate the predictive performances on the test set
mse = mean_squared_error(y_test, test_preds)
rmse = sqrt(mse)
print("the rmse result for the testing data is :", rmse)
```
