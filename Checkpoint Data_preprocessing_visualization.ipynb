```python
import matplotlib.pyplot as plt 
import seaborn as sns

def plot_correlation_map( df ):
    corr = df.corr()
    plt=__import__("matplotlib.pyplot")
    s,ax= plt.pyplot.subplots(figsize =( 12 , 10 ) )
    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )
    s = sns.heatmap(corr, cmap = cmap, square=True, cbar_kws={'shrink' : .9}, ax=ax, annot = True, annot_kws = {'fontsize': 12 })
```

```python
plot_correlation_map(df)
```

This fuction is used as a tool to correlate between different features in a dataset. It is represented as a correlation matrix. It allows us to visualize the degree of correlation that exists between different variables for example the correlation between (Fare, Pclass) has the lowest degree and the value is -0.55 whereas the same features (Age,Age) have the highest degree of correlation which is 1. I think this function will allow us to reduce the number of features and to avoid the useless columns. 
This is a dimensionality reduction and it is used to improve the runtime and effectiveness of this model.
